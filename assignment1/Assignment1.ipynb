## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
import re, collections

class BPETokenizer:
    """
    Byte Pair Encoding Tokenizer. (https://arxiv.org/pdf/1508.07909.pdf)

    Attributes:
        vocab_size (int): The size of the vocabulary.
        vocab (set): The set of unique symbols in the vocabulary.
        splits (dict): Dictionary to store split symbols and their frequencies.
        merges (dict): Dictionary to store merged pairs.
    """

    def __init__(self, vocab_size: int) -> None:
        """
        Initialize the BPETokenizer.

        Args:
            vocab_size (int): The size of the vocabulary.
        """
        self.vocab_size = vocab_size
        self.vocab = set()
        self.splits = {}
        self.merges = {}

    # Function from https://arxiv.org/pdf/1508.07909.pdf
    def get_stats(self, splits):
        """
        Calculate statistics of symbol pairs in the splits.

        Args:
            splits (dict): Dictionary containing split symbols and their frequencies.

        Returns:
            dict: Dictionary containing symbol pairs and their frequencies.
        """
        # Initialize dictionary to store symbol pairs and their frequencies
        pairs = collections.defaultdict(int)
        for word, freq in splits.items():
            symbols = word.split()
            for i in range(len(symbols) - 1):
                # Increment the frequency count for the symbol pair
                pairs[symbols[i], symbols[i + 1]] += freq
        return pairs

    # Function from https://arxiv.org/pdf/1508.07909.pdf
    def merge_pair(self, pair, v_in):
        """
        Merge the given symbol pair in the vocabulary.

        Args:
            pair (tuple): The symbol pair to be merged.
            v_in (dict): Dictionary containing split symbols and their frequencies.

        Returns:
            dict: Updated dictionary after merging the symbol pair.
        """
        v_out = {}
        bigram = re.escape(' '.join(pair))
        # Compile regular expression pattern to match the symbol pair
        p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        for word in v_in:
            # Substitute the symbol pair with its merged form in the word
            w_out = p.sub(''.join(pair), word)
            v_out[w_out] = v_in[word]
        return v_out


    def initialize_splits(self, corpus: List[str]):
        """
        Initialize the splits dictionary from the given corpus.

        Args:
            corpus (List[str]): List of strings representing the corpus.

        Returns:
            dict: Dictionary containing split symbols and their frequencies.
        """
        splits = {}
        for text in corpus:
            text = text.lower()
            # Replace multiple whitespaces with a single space
            text = re.sub(r'\s+', ' ', text)
            all_words = text.split()
            for word in all_words:
                # Convert each character in the word to a symbol and append '</w>'
                word = ' '.join(list(word)) + ' ' + '</w>'
                # Increment the frequency count of the word in the splits dictionary
                splits[word] = splits.get(word, 0) + 1
        return splits


    def initialize_vocab(self, splits):
        """
        Initialize the vocabulary from the splits.

        Args:
            splits (dict): Dictionary containing split symbols and their frequencies.

        Returns:
            set: Set of unique symbols in the vocabulary.
        """
        vocab = set()
        for word in splits:
            for s in word:
                vocab.add(s)
        return vocab


    def train(self, corpus: List[str]) -> None:
        """
        Train the tokenizer on the given corpus.

        Args:
            corpus (List[str]): List of strings representing the corpus.
        """
        # Initialize splits dictionary and vocabulary from the corpus
        self.splits = self.initialize_splits(corpus)
        self.vocab = self.initialize_vocab(self.splits)

        while len(self.vocab) < self.vocab_size:
            pairs = self.get_stats(self.splits)
            # Find the most frequent symbol pair
            best = max(pairs, key=pairs.get)
            # Merge the most frequent symbol pair in the splits
            self.splits = self.merge_pair(best, self.splits)
            # Add the merged symbol pair to the vocabulary
            self.vocab.add(''.join(list(best)))
            self.merges[best] = ''.join(list(best))


    def split_words(self, text: str) -> List[str]:
        """
        Split words into characters and add end-of-word token.

        Args:
            text (str): Input text.

        Returns:
            List[str]: List of split symbols with end-of-word token.
        """
        words = []
        for word in text.lower().split():
            # Replace multiple whitespaces with a single space
            word = re.sub(r'\s+', ' ', word)
            # Convert each character in the word to a symbol and append end-of-word token
            word = ' '.join(list(word)) + ' ' + '</w>'
            words.append(word)
        return words


    def get_pairs(self, words: List[str]):
        """
        Get symbol pairs from the list of split symbols.

        Args:
            words (List[str]): List of split symbols.

        Returns:
            List[tuple]: List of symbol pairs.
        """
        pairs = []
        for word in words:
            symbols = word.split()
            for i in range(len(symbols) - 1):
                # Append the symbol pair to the pairs list
                pairs.append((symbols[i], symbols[i + 1]))
        return pairs


    def merge_pair_token(self, pair, v_in):
        """
        Merge the given symbol pair in the list of split symbols.

        Args:
            pair (tuple): The symbol pair to be merged.
            v_in (List[str]): List of split symbols.

        Returns:
            List[str]: Updated list after merging the symbol pair.
        """
        v_out = []
        # Escape the symbol pair to handle special characters in regular expression
        bigram = re.escape(' '.join(pair))
        # Compile regular expression pattern to match the symbol pair
        p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        for word in v_in:
            # Substitute the symbol pair with its merged form in the word
            w_out = p.sub(''.join(pair), word)
            v_out.append(w_out)
        return v_out


    def tokenize(self, text: str) -> List[str]:
        """
        Tokenize the input text.

        Args:
            text (str): Input text.

        Returns:
            List[str]: List of tokens.
        """
        tokens = []
        words = self.split_words(text)
        # Get the number of merge operations
        count_merges = len(self.merges)
        while count_merges != 0:
            flag_merges = 0
            # Get symbol pairs from the split words
            pairs = self.get_pairs(words)
            # Checking that there are still pairs
            if len(pairs) == 0:
                break
            for pair in pairs:
                if pair in self.merges:
                    words = self.merge_pair_token(pair, words)
                    count_merges -= 1
                    flag_merges += 1
            # Checking to see if there are any pairs that can be merged.
            if flag_merges == 0:
                break

        for word in words:
            token = word.split()
            tokens += token

        return tokens
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000].
```python
import matplotlib.pyplot as plt


vocab_sizes = [len(set(''.join(corpus))), 250, 500, 750, 1000]
avg_lengths = []
```

```python
plt.plot(vocab_sizes, avg_lengths, marker='o')
plt.xlabel('Vocabulary size')
plt.ylabel('Average length of tokenes')
plt.title('Dependency of tokenes length on vocabulary size')
plt.show()
```

I think vocab_size 500 is preferable as it looks more balanced. To can see that as vocab_size increases, the average number of tokens decreases because it processes and merges more pairs.
